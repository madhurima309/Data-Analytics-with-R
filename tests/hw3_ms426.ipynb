```r
IBM = read_csv(file = 'IBM.csv', col_types = cols())
MSFT = read_csv(file = 'MSFT.csv',col_types = cols())
GOOG = read_csv(file = 'GOOG.csv', col_types = cols())
AAPL = read_csv(file = 'AAPL.csv', col_types = cols())
AMZN = read_csv(file = 'AMZN.csv', col_types = cols())
FB = read_csv(file = 'FB.csv', col_types = cols())
```

```r
print(IBM)
```

```r
DIV = function(x){
    dtset=x[,c(1,5,6)]
    dtset$dffr1=lag(dtset$Close,1)/dtset$Close
    dtset$dffr2=lag(dtset$'Adj Close',1) / dtset$'Adj Close'
    dtset=mutate(dtset,dffr=dtset$dffr1-dtset$dffr2)
    dtset=mutate(dtset,Divi=dtset$dffr*dtset$Close)
    dtset=subset(dtset,(Divi!='0'))
    dtset[,c(1,7)]
}
```

```r
DIV1 = function(x){
    dtset=x[,c(1,5,6)]
    dtset$dffr1=lag(dtset$Close,1)/dtset$Close
    dtset$dffr2=lag(dtset$'Adj Close',1) / dtset$'Adj Close'
    dtset=mutate(dtset,dffr=dtset$dffr1-dtset$dffr2)
    dtset=mutate(dtset,Divi=dtset$dffr*dtset$Close)
    dtset[,c(1,7)]
}
```

```r
Div_IBM = DIV(IBM)
Div_MSFT = DIV(MSFT)
Div_GOOG = DIV(GOOG)
Div_AAPL = DIV(AAPL)
Div_AMZN = DIV(AMZN)
Div_FB = DIV(FB)
```
